## Preparing the Data

Now that we have a dataset with tweets that are in english and have at least 10 likes  we will start by preparing our training and testing datasets. Tweets posted between 1/1/2016 to 1/10/2018 will be part of the training dataset and the tweets posted between 1/3/2019 to 23/11/2019 will be our testing dataset.

```python
import pandas as pd
import numpy as np
```

```python
df_tweets = pd.read_csv('tweets_processed.csv',parse_dates=True,
                        date_parser=lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S+00'), 
                        index_col='timestamp')
```

```python
df_tweets
```

```python
#Let's start by deleting the columns that are irrelevant.

df_tweets = df_tweets.drop(columns=['Unnamed: 0','index','language'])
```

```python
df_tweets = df_tweets.rename(columns = {'text\r':'tweet'})
df_tweets
```

```python
df_tweets = df_tweets.resample('D').agg({'replies': sum, 'likes': sum, 'retweets': sum, 'tweet': ' '.join})
```

```python
df_tweets
```

```python
# remove the tweet posted in 2009 as it was the first tweet posted about Bitcoin and there are no tweets available between 2009 and 1/1/2016
# we will also remove the days without a single tweet.

df_tweets = df_tweets[df_tweets.index > pd.Timestamp(2016,1,1)]
df_tweets = df_tweets[df_tweets['replies'] > 0]
```

```python
df_tweets
```

```python
df_tweets.plot()
```

## Comments

We can see an increase in popularity in July 2017 and another one around March 2019. Now that the dataset is resampled with a daily timestep we can split the data into a training and testing datasets. Approximately 70% of the data will be used to train the model and 30% of the data will be used to test the model.

```python
data_train = df_tweets[df_tweets.index < pd.Timestamp(2019,3,1)]
data_test = df_tweets[df_tweets.index >= pd.Timestamp(2019,3,1)]
```

## Preparing the Labels

We already have the training and testing dataset but without the label (i.e. if the price of Bitcoin went up or down). We will use the second dataset to get this information.

```python
df_price = pd.read_csv('datasets/COIN_BTC.csv',parse_dates=True,
                        date_parser=lambda x: pd.datetime.strptime(x, '%d/%m/%Y'), 
                        index_col='date')
```

```python
df_price
```

As we want to be able to predict what will happen the next day, we can use the percent_change_24h the next day to know if for a given day the price went up or down

```python
df_price['Next Day'] = df_price['percent_change_24h'].shift(-1)
```

```python
df_price  
```

```python
def up_or_down(row):
    if row['Next Day'] > 0:
        return 1
    else:
        return 0
    
df_price['Next Day'] = df_price.apply(up_or_down, axis = 1)
```

```python
df_price
```

```python
# Now we can create the label for our training and testing datasets

label_train = df_price.loc[data_train.index,'Next Day']
label_test = df_price.loc[data_test.index,'Next Day']
```

# Processing the Data

Now we will process the data and follow standard NLP processing techniques. (convert the text to lower case, keep only alpha numeric characters, remove stopwords, stemming and lemmatization, bag of words (based on the training dataset only)). Once all these steps are achieved we will reformat our data_training and data_testing dataframe accordingly before moving to the classification model.

```python
# We will process the tweets, the same way we processed the movie reviews for the mini project at the start of this course.

import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
import re
from bs4 import BeautifulSoup

def review_to_words(tweet):
    text = BeautifulSoup(tweet, "html.parser").get_text() # Remove HTML tags
    text = re.sub(r"[^a-zA-Z0-9]", " ", text.lower()) # Convert to lower case
    words = text.split() # Split string into words
    words = [w for w in words if w not in stopwords.words("english")] # Remove stopwords
    words = [PorterStemmer().stem(w) for w in words] # stem
    
    return words
```

```python
def preprocess_data(data_train, data_test):
    """Convert daily tweets to words"""

    words_train = [review_to_words(tweet) for tweet in data_train['tweet']]
    words_test = [review_to_words(tweet) for tweet in data_test['tweet']]

    
    return words_train, words_test
```

```python
# Preprocess data
words_train, words_test = preprocess_data(data_train, data_test)
```

## Bag of Word

The next step is to create a bag of word based on the training dataset

```python
from sklearn.feature_extraction.text import CountVectorizer

def extract_BoW_features(words_train, words_test, vocabulary_size=2000):

    vectorizer = CountVectorizer(max_features=vocabulary_size,
            preprocessor=lambda x: x, tokenizer=lambda x: x)
    features_train = vectorizer.fit_transform(words_train).toarray()

    # Apply the same vectorizer to transform the test documents (ignore unknown words)
    features_test = vectorizer.transform(words_test).toarray()
    vocabulary = vectorizer.vocabulary_
    
    features_train = np.append(features_train,data_train.loc[:,['likes','replies','retweets']].values, axis = 1)
    features_test = np.append(features_test,data_test.loc[:,['likes','replies','retweets']].values, axis = 1)   

    return features_train, features_test, vocabulary
```

```python
# Extract Bag of Words features for both training and test datasets
train_X, test_X, vocabulary = extract_BoW_features(words_train, words_test, vocabulary_size=5000)
```

Three model were considered:

1. Random Forest Classifier
2. XGB
3. Neural Network

After cleaning the data and merging all the tweets of the same day. We ended up with a relatively small dataset, therefore, I will only test the random forest classifier and XGB. I will start by testing the two model and optimize the most promising one.

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(max_depth=10, random_state=0)
RF.fit(train_X, label_train)
```

```python
from sklearn.metrics import accuracy_score

y_predict = RF.predict(test_X)
accuracy_score(label_test, y_predict)
```

```python
from xgboost import XGBClassifier
XGB = XGBClassifier()
XGB.fit(train_X, label_train)
```

```python
y_predict = XGB.predict(test_X)
accuracy_score(label_test, y_predict)
```

## Comments

Both model gave similar result, the random forest model is marginally better. Both model are performing worse than the benchmark model. The next step will be to try to tune the random forest parameters to improve it's performance.

```python
from sklearn.model_selection import GridSearchCV

parameters = {'bootstrap': [True, False],
              'n_estimators': [5,10,20,50,100,200,500,1000],
             'max_depth':[3,5,8,12,20,50,100],
             'min_samples_split':[2,3,4],
             'max_features':['sqrt', 'log2']}

RF = RandomForestClassifier()

model = GridSearchCV(RF, parameters)

model.fit(train_X, label_train)
```

```python
model.best_params_
```

```python
y_predict = model.predict(test_X)
accuracy_score(label_test, y_predict)
```
